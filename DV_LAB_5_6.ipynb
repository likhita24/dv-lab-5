```python
plt.figure(figsize = (8,8))
EDA = adult_ds['income'].value_counts(normalize = True).rename_axis('income').reset_index(name = 'Percentage')
area = sns.barplot(x = 'income', y = 'Percentage', data = EDA.head(10), palette='mako')

plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(1))
for p in area.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    area.annotate(f'{height:.0%}', (x + width/2, y + height*1), ha='center')
```

```python
def agebargraphs(arr):
    arr = int(arr)
    arr = abs(arr)
    if( 18 < arr < 31 ):
        return "19-30"
    if( 30 < arr < 41 ):
        return "31-40"
    if( 40 < arr < 51 ):
        return "41-50"
    if( 50 < arr < 61 ):
        return "51-60"
    else:
        return "Greater than 60"

adult_ds['age_group'] = adult_ds['age'].apply(agebargraphs)
plt.figure(figsize=(12,12))
ages = ['19-30', '31-40', '41-50', '51-60', 'Greater than 60']
sns.countplot(adult_ds['age_group'], hue = adult_ds['income'], palette='mako', order = ages)
```
